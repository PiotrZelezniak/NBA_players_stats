```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

r=requests.get("https://www.basketball-reference.com/leagues/NBA_2022_per_game.html")
c=r.content

soup=BeautifulSoup(c,"html.parser")
print(soup)
```

```python
header=soup.find('div',{'id':'div_per_game_stats'})
columns=[h.text for h in header.find_all('th',{'scope':'col'})]
pd_header=pd.DataFrame(columns=columns[1:30])
pd_header
```

```python
players=soup.find_all('tr',{'class':'full_table'})
df=pd.Series()
for player in players:
    stats=[stat.text for stat in player.find_all('td')]
    temp_df=pd.DataFrame(stats).transpose()
    temp_df.columns=columns[1:30]
    df=pd.concat([df,temp_df],ignore_index=True)
df=df.drop(df.columns[0],axis=1)
df
```
